```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('class_data.csv')
df.head()
```

```python
df['instagram_followers'].name = 'Amount of instagram followers' #Labeling the variable for plotting
file_size = df['instagram_followers'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(file_size, kde=False) 
plt.title('Instagram followers')
plt.show()
```

```python
sns.boxplot(df['apps_phone'])
plt.title('Apps on phone')
plt.xlabel('Amount of apps on phone')
plt.show()
```

```python
df['gender'].value_counts().plot(kind='bar');
```

```python
df_paid = df[df['linkedin_con'] > 0] #Subsetting the dataframe to include only paid apps
sns.violinplot(df_paid['linkedin_con'])
sns.swarmplot(df_paid['linkedin_con'], color='purple') #default color doesn't stand out
plt.title('Linkedin followers')
plt.ylabel('amount')
plt.xlabel('followers')
plt.show()
```

```python
sns.distplot(df['texts_yesterday'])
plt.title('Texts yesterday')
plt.ylabel('frequency')
plt.xlabel('Amount of texts yesterday')
plt.show() 
```
